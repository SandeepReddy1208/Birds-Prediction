IMPLEMENTATION OF YOLO WITH IMAGE DETECTION OF GROUP BIRDS

```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from collections import defaultdict
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from ultralytics import YOLO

# Load the trained MobileNet CNN model for bird classification
model = load_model("model_RMSprop.h5")

# Load YOLOv8 model for object detection (pre-trained on COCO dataset)
yolo_model = YOLO("yolov8n.pt")  # Using the small YOLOv8 model

# Define class labels (same as during training)
valid_dir = "S:/Labs/dl/Birds_25/valid"  # Path to validation directory
class_labels = sorted(os.listdir(valid_dir))

# Input single bird image
image_path = "C:/Users/Sandeep/Downloads/generated_birds.jpeg"  # Change this to your image path
output_folder = "S:/Labs/dl/Birds_Output/"
os.makedirs(output_folder, exist_ok=True)

# Dictionary to store cumulative class probabilities
class_probabilities = defaultdict(float)

# Read the input image
frame = cv2.imread(image_path)
if frame is None:
    print("Error: Unable to read the image.")
    exit()

# Resize the frame for better display
frame = cv2.resize(frame, (800, 800))

# Use YOLO to detect objects in the image
results = yolo_model(frame)

for result in results:
    for box in result.boxes.xyxy:  # Get bounding boxes
        x1, y1, x2, y2 = map(int, box)
        label = result.names[int(result.boxes.cls[0].item())]
        
        if label == "bird":  # Only process birds
            bird_crop = frame[y1:y2, x1:x2]
            bird_crop = cv2.resize(bird_crop, (224, 224))
            img_array = image.img_to_array(bird_crop) / 255.0  # Normalize
            img_array = np.expand_dims(img_array, axis=0)

            # Predict bird species
            pred_probs = model.predict(img_array)[0]
            pred_class = np.argmax(pred_probs)
            predicted_label = class_labels[pred_class]
            confidence = pred_probs[pred_class] * 100

            # Store probabilities for graph
            class_probabilities[predicted_label] += confidence  

            # Draw bounding box and label
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            label_text = f"{predicted_label}: {confidence:.2f}%"
            cv2.putText(frame, label_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)

# Save and display the processed image
output_path = os.path.join(output_folder, "output.jpg")
cv2.imwrite(output_path, frame)
cv2.imshow("Bird Detection", frame)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Sort class probabilities in descending order
sorted_probs = sorted(class_probabilities.items(), key=lambda x: x[1], reverse=True)

# Extract class names and summed probabilities
if sorted_probs:
    classes, probs = zip(*sorted_probs)
    
    # Normalize probabilities for visualization
    probs = np.array(probs) / sum(probs) * 100  # Convert to percentage

    # Plot class resemblance graph
    plt.figure(figsize=(10, 5))
    plt.barh(classes, probs, color="skyblue")
    plt.xlabel("Resemblance (%)")
    plt.ylabel("Bird Species")
    plt.title("Bird Class Resemblance in Image")
    plt.gca().invert_yaxis()
    plt.show()

    # Print the most likely bird species
    most_likely_bird = classes[0]
    print(f"\nThe most likely bird species in the image is: {most_likely_bird} ({probs[0]:.2f}% resemblance)")
else:
    print("No birds detected in the image.")
```
